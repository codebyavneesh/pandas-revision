```python
def is_prime(val):
    try:
        end=int(val/2)+1
    except ValueError:
        return False

    for i in range(2, end):
        if val%i==0:
            return False
    return True
```

```python
is_prime(5)
```

```python
mask=df.Age.apply(is_prime)
df[mask]
```

```python
def lt25(val):
    return val<25
mask1=df.Age.apply(lt25)
df[mask1]
```
